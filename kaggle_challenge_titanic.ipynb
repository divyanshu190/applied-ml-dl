```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("challenge_titanic_kaggle_data_set.csv")
df
```

```python
# column which have no role in training/survival :-> Name, Ticket, Fare, Cabin, PassengerID
```

```python
column_to_drop = ['Name','Ticket','Cabin','PassengerId']
df.drop(column_to_drop, axis = 1, inplace = True)
df
# name wala column remove kar diya 
```

```python
df.isnull().sum()
```

```python
# Age :-> average age daal denga 
# cabin :-> lable encoder, but ya pehle use fill bhi karna h toh apple se fill kar dunga
# embarked ;-> lable encoder, same here fill with something like poco 
```

```python
# age done
av_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(av_age)

# embarked
df['Embarked'] = df['Embarked'].fillna("poco")
df.isnull().sum()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
lable1 = LabelEncoder()
#lable2 = LabelEncoder()
lable3 = LabelEncoder()
df['Sex']= lable1.fit_transform(df['Sex'])
#df['Cabin']= lable2.fit_transform(df['Cabin'])
df['Embarked']= lable3.fit_transform(df['Embarked'])
df
```

```python
# model choosing 
# decision tree, adaboost
# adaboost is better then DT usme bhi uska tuned version 
```

```python
x = df[["Pclass", "Sex", "Age", "SibSp", "Parch","Fare", "Embarked"]]
y = df['Survived']
#x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.6)
```

```python
# this is the code for adaboost algo which had given a accuracy of = 0.8518518518518519
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
model = AdaBoostClassifier(
    estimator = DecisionTreeClassifier(),
    random_state = 42
)

parameter_grid = {
    "n_estimators":[50, 100, 200, 300, 500],
    "learning_rate":[0.01, 0.1, 0.2, 0.5, 1.0],
    "estimator__max_depth":[1, 2, 3, 4],
    "algorithm":["SAMME", "SAMME.R"]
}

from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(
    estimator = model,
    param_grid = parameter_grid,
    cv = 5,
    scoring = "accuracy",
    n_jobs =-1
)
grid.fit(x,y)
best_model = grid.best_estimator_

apple = best_model.predict(x)
from sklearn.metrics import accuracy_score

print(accuracy_score(apple,y))
```

```python
from sklearn.linear_model import LogisticRegression
best_model = LogisticRegression(max_iter=1000)
best_model.fit(x,y)
apple = best_model.predict(x)
from sklearn.metrics import accuracy_score
print(accuracy_score(apple,y))
```

```python
test = pd.read_csv("challenge_titanic_kaggle_test_set.csv")
passenger_ids = test["PassengerId"]
column_to_drop = ['Name', 'Ticket','Cabin','PassengerId']
test.drop(column_to_drop, axis = 1, inplace = True)

av_age = test['Age'].mean()
test['Age'] = test['Age'].fillna(av_age)

# cabin
#test['Cabin'] = test['Cabin'].fillna("apple")

av_fare = test['Fare'].mean()
test['Fare'] = test['Fare'].fillna(av_fare)

test['Embarked'] = test['Embarked'].fillna("poco")
test.isnull().sum()


lable4 = LabelEncoder()
lable6 = LabelEncoder()
test['Sex']= lable4.fit_transform(test['Sex'])
test['Embarked']= lable6.fit_transform(test['Embarked'])
answer = best_model.predict(test)
```

```python
final = pd.DataFrame({
    "PassengerID": passenger_ids,
    "Survived": answer
})
final.to_csv("69new_solution_titanic_kaggle.csv", index=False) 
```
