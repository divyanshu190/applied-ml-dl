```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("dataset_for_ANN.csv")
```

```python
df.drop(columns=[df.columns[0], df.columns[1]], inplace= True)
```

```python
to = ['Geography', 'Gender', 'Surname']
df[to] = df[to].apply(LabelEncoder().fit_transform)
```

```python
#this piece of code is for standard scaler 
col = ['CreditScore', 'Balance', 'EstimatedSalary', 'Age', 'Surname']
for i in col:
    mean = df[i].mean()
    std = df[i].std()
    df[i] = (df[i]-mean)/std
```

```python
x = df.drop("Exited", axis=1)
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
y_train = y_train.to_numpy().reshape(-1, 1)
y_test = y_test.to_numpy().reshape(-1, 1)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
```

```python
model.add(Input(shape=(x_train.shape[1],)))
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.3))
```

```python
model.add(Dense(units=64, activation='relu'))
model.add(Dropout(0.3))
```

```python
model.add(Dense(units=32, activation='relu'))
```

```python
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
early_stopper = EarlyStopping(
    monitor = 'val_loss',
    patience = 10,
    restore_best_weights = True
)
```

```python
history = model.fit(
    x_train, y_train,
    epochs = 100,
    batch_size = 32,
    validation_split = 0.2,
    callbacks = [early_stopper],
    verbose = 1
)
```

```python
loss, apple = model.evaluate(x_test, y_test)
print(apple)
print(loss)
```
